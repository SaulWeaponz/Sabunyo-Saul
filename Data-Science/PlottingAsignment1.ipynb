```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('data.csv')
# Displaying some of the first few rows of the dataset
df.head(5)
```

```python
 #First: Data Preprocessing or Data Cleaning before plotting
# 1.Handling  Empty values
# 2.Handling Wrong formats   
# 3. Handling Wrong data     
# 4.Handling dataset with Duplicates  


# 1.Handling  Empty values
# Checking for empty values
print(df.isnull().sum())
```

```python
# Fill empty values with mean 
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Check for non numeric values in numeric columns
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    if not pd.to_numeric(df[col], errors='coerce').notnull().all():
        print(f"Non-numeric values found in column: {col}")
```

```python
#Wrong data
# Check for negative values in numeric columns
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    if (df[col] < 0).any():
        print(f"Negative values found in column: {col}")
```

```python
# Duplicates
duplicates = df.duplicated()
if duplicates.any():
    print("Duplicates found:")
    print(df[duplicates])
else:
    print("No duplicates found.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up figure size
plt.figure(figsize=(15, 10))
#Histogram for Maxpulse
plt.subplot(2,2,1)
sns.histplot(df['Maxpulse'], bins=30, kde=True)
plt.title('Maxpulse Distribution')

#Histogram for Maxpulse
plt.subplot(2,2,2)
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Calories Distribution')
```

```python
# Scatter plot of 'Maxpulse' Vs 'Calories'
# Setting up figure size
plt.Figure(figsize=(15, 10))
plt.subplot(2,2,1)
sns.scatterplot(x='Maxpulse', y='Calories', data=df)
plt.title('Maxpulse Vs Calories')

# Scatter plot of  'Calories' Vs 'Pulse'
plt.subplot(2,2,2)
sns.scatterplot(x='Calories', y='Pulse', data=df)
plt.title('Calories Vs Pulse')
```
